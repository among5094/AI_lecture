```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
#sklearn: 머신러닝 전용 라이브러리
```

```python
#유방암데이터 불러오기
cancer = datasets.load_breast_cancer()

df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
```

```python
df.head()
# target -> 0이 앙성, 1이 악성
```

```python
#우리가 사용할 input데이터와 output데이터로 분류
input = cancer.data
output=df['target'] #아웃풋은 양성은지 음성인지
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_output, test_output = train_test_split(input,output, test_size=0.2)
#이상태는 train 75%, test 25%로 들어감 -> 그래서 test_size옵션 사용하는 거임
# 0.2는 8:2로 나눌 수 있음. 7.5대 2.5도 사용됨
```
